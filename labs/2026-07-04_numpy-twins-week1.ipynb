If the plot surprises you, that surprise is the lesson: chase it.

Twins in this lab: the Bangalore house drill (covariance → eigen → perpendicular axes)
and the two-drive latency drill (same mean, different σ).

```python
# Twin 1 — the §21 house drill: covariance cloud + PERPENDICULAR eigen-axes
import numpy as np, matplotlib.pyplot as plt

X = np.array([[3,4],[4,7],[6,5],[6,7],[6,7]], float)   # 5 houses: School, Metro
C = np.cov(X.T); lam, V = np.linalg.eig(C)
print('C =\n', C, '\nlambda =', lam, '  v1.v2 =', V[:,0] @ V[:,1])

m = X.mean(0)
plt.figure(figsize=(5,5))
plt.scatter(*X.T, s=110, zorder=3, label='5 houses')
for i, c in zip(np.argsort(lam)[::-1], ['tab:orange','tab:green']):
    v = V[:, i] * np.sqrt(lam[i]) * 1.6
    plt.arrow(*m, *v, head_width=.14, color=c, lw=2.2,
              length_includes_head=True, label=f'eigen-axis  λ={lam[i]:.0f}')
plt.xlabel('School score'); plt.ylabel('Metro score')
plt.axis('equal'); plt.grid(alpha=.3); plt.legend()
plt.title('Covariance cloud + perpendicular eigen-axes');
```

```python
# Twin 1b — the what-if: H3 (6,5) -> (6,9). Same code, one number changed.
X2 = np.array([[3,4],[4,7],[6,9],[6,7],[6,7]], float)
C2 = np.cov(X2.T); lam2, V2 = np.linalg.eig(C2)
print('C =\n', C2, '\nlambda =', lam2, '  v1.v2 =', V2[:,0] @ V2[:,1])
m2 = X2.mean(0)
plt.figure(figsize=(5,5))
plt.scatter(*X2.T, s=110, zorder=3, label='5 houses (H3 moved)')
for i, c in zip(np.argsort(lam2)[::-1], ['tab:orange','tab:green']):
    v = V2[:, i] * np.sqrt(lam2[i]) * 1.2
    plt.arrow(*m2, *v, head_width=.14, color=c, lw=2.2,
              length_includes_head=True, label=f'λ={lam2[i]:.2f}')
plt.xlabel('School'); plt.ylabel('Metro'); plt.axis('equal'); plt.grid(alpha=.3); plt.legend()
plt.title('Was your prediction right? Perpendicular still? (check v1.v2)');
```

### 🔮 Predict
Two drives, both mean = 100 µs. A: [98,100,102,99,101] · B: [80,120,90,110,100].
Roughly what σ do you expect for each? (You computed these by hand: ~1.6 vs ~15.8.)
Which bell will be the razor spike?

```python
# Twin 2 — same mean, different sigma: the bells expose the lie
A = np.array([98,100,102,99,101], float)
B = np.array([80,120,90,110,100], float)
for nm, d in [('A', A), ('B', B)]:
    print(f'Drive {nm}: mean={d.mean():.0f}  sigma={d.std(ddof=1):.2f}')

x = np.linspace(60, 140, 400)
bell = lambda x, mu, s: np.exp(-(x-mu)**2/(2*s*s)) / (s*np.sqrt(2*np.pi))
plt.figure(figsize=(7,4))
plt.plot(x, bell(x, A.mean(), A.std(ddof=1)), lw=2, label=f'A  σ={A.std(ddof=1):.2f} (steady)')
plt.plot(x, bell(x, B.mean(), B.std(ddof=1)), lw=2, label=f'B  σ={B.std(ddof=1):.2f} (jitter-bomb)')
plt.scatter(A, np.zeros_like(A)-0.004, marker='|', s=200, label='A readings')
plt.scatter(B, np.zeros_like(B)-0.010, marker='|', s=200, label='B readings')
plt.xlabel('latency (µs)'); plt.ylabel('density'); plt.legend(); plt.grid(alpha=.3)
plt.title('Identical means — σ is the whole story');
```
